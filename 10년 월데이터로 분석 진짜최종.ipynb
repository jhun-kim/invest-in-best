```python
# 코스피 수익률 분산 (36개월 = 3년치 분산)
import numpy as np
index_var = []
for i in range(36):
    index_var.append(0)

for i in range(len(df_2.index)-36):
    index_var.append(np.var(df_2['코스피200수익률'].iloc[i:i+36]))

df_2['코스피분산'] = index_var
```

```python
#코스피와 개별주가 공분산 구하기 3년 / 월단위
index_covar = []

for i in range(36):
    index_covar.append(0)

for i in range(len(df_2.index)-36):
    a=df_2['코스피200수익률'].iloc[i:i+36]
    b=df_2['수익률'].iloc[i:i+36]
    index_covar.append(a.cov(b))

index_covar
```

```python
df_2["공분산"] = index_covar
```

```python
df_2["베타"]=df_2["공분산"]/df_2['코스피분산']
```

```python
#CAPM 기대수익률 계산 

df_2["CAPM 기대수익률"] = df_2['무위험월기하평균']+ df_2['베타']*(df_2['코스피200수익률']-df_2['무위험월기하평균'])
```

```python
df_2.reset_index(inplace=True)
df_2
```

```python
#가중치 조절하기
weight = [1,1,1,1]
n =0
for i in range(len(weight)):
    n += weight[i]
n
```

```python
c=df_2["거래소코드"].unique()
df_2["가중치CAPM"] = df_2["CAPM 기대수익률"]
df_2["가중치무위험월기하평균"] = df_2["무위험월기하평균"]
for i in range(len(df_2)):
    for j in range(len(c)):
        if df_2["거래소코드"][i]==c[j]:
            df_2["가중치CAPM"][i] = df_2["CAPM 기대수익률"][i] * weight[j]
            df_2["가중치무위험월기하평균"][i] = df_2["무위험월기하평균"][i] * weight[j]

```
