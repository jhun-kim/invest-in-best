```python
def three_pass(a,b,c):
  if  a+b+c == 3 :
    result = 1

  else:
    result = 0

  return result

def four_pass(a,b,c,d):
  if  a+b+c+d == 4 :
    result = 1

  else:
    result = 0

  return result

def five_pass(a,b,c,d,e):
  if  a+b+c+d+e == 5 :
    result = 1

  else:
    result = 0

  return result

def six_pass(a,b,c,d,e,f):
  if  a+b+c+d+e+f == 6 :
    result = 1

  else:
    result = 0

  return result
```
